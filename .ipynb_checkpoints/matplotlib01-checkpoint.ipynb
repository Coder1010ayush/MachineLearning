### Drawing a line
---



```python
import matplotlib.pyplot as plt
import numpy as np
arrx = np.array([1,2,3,4,5])
arry = np.array([10,20,30,40,50])
plt.plot(arrx,arry,linewidth =2,color='b',linestyle='--')
plt.xlabel(" x axis ")
plt.ylabel(" y axis ")
plt.show()
```

### Drawing a line using more parameters

```python
import matplotlib.pyplot as plt
import numpy as np
import sys
fruits = np.array(["apple","orange","banana","guava","grapes"])
price = np.array([100,200,130,400,320])

plt.plot(fruits,price)
plt.xlabel("Fruits",color='g')
plt.ylabel("Price",color='b')
plt.show()

plt.plot(fruits,price,'o') # it will only shows the points on the of instersection
plt.xlabel("Fruit",color='g')
plt.ylabel("Price of fruit",color='b')
plt.show()
```

### Representing Points in a 2D Plane

```python
import matplotlib.pyplot as plt
import numpy as np

number_of_toys = np.array([2,4,6,8,10])
price_of_toys = np.array([100,200,300,400,500])
plt.plot(number_of_toys,'r')

plt.plot(price_of_toys,'g')
plt.plot(number_of_toys,price_of_toys)
plt.xlabel("number of toys")
plt.ylabel("price of toys")

plt.show()
```

### Markers and its parameters

```python
import matplotlib.pyplot as plt
import numpy as np
import sys

numbers = np.array([1,2,3,4,5])
ferquency = np.array([10,20,30,40,50])

plt.plot(numbers,ferquency,marker='o',ms =20,mec='r',mfc='g') # there is a shorthand marker line color in single inverted comma

plt.xlabel("number")
plt.ylabel("frequency")
plt.show()
```

### Line Styling in matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import sys

heat = np.array([100.3,150.8,98.6,67.8,135.6])
vibration = np.array([20,35,17,14,30])
plt.plot(vibration,heat,'o')

plt.ylabel("heat")
plt.xlabel("vibration")

plt.show()

```

```python
integer = np.array([1,2,3,4,5,6,7])
square = np.array([1,4,9,16,25,36,49])
plt.plot(integer,square,marker='o',linestyle='dashed',linewidth='0.7')
plt.xlabel("integer")
plt.ylabel("square")
plt.show()
```

###Labels of the graphical representation

```python
import matplotlib.pyplot as plt
import numpy as np
import sys

books = np.array(["Java","Network Synthesis","Statistics","Probability"])
price = np.array([980,1290,450,670])
font = {'family':'serif','color':'blue','size':12}
plt.plot(books,price,marker='o')

plt.title("Books and price graph",loc='center',color='r',fontdict = font)
plt.xlabel("books")
plt.ylabel("price")

plt.show()
```
