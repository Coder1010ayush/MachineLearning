```python
import numpy as np
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
import pandas as pd
import math
```

```python
class Standardization:
    def __init__(self):
        pass
    
    def mean_of_list_element(self,df):
        group = df.to_numpy(dtype='float32')
        sum=0
        for i in group:
            sum += i;
        return sum/len(group)
    
    def mean_of_list_elements(self,df):
        group = df.to_numpy(dtype='float16')
        return group.sum(dtype='float16')/len(group)
    
    def median_of_list_elements(self,df):
        group = df.to_numpy(dtype='float16')
        group = np.sort(group)
        size = np.size(group)
        if(np.size(group) %2==0):
            
            f1 = group[int(size/2)]
            f2 = group[int(size/2)+1]
            median = float((f1+f2)/2)
            return median
        else:
            median = group[int(size/2)]
            return median
    
    def standard_deviation_list_elements(self,df):
        mean = self.mean_of_list_elements(df)
        group = df.to_numpy(dtype='float16')
        sum=0
        for i in range(len(df)):
            x=mean-group[i]
            sum += x*x
        size = np.size(group)
        return math.sqrt(sum/size)
    
    def replace_df_to_standard(self,df):
        local = df.to_numpy(dtype='float16')
        mean = self.mean_of_list_elements(df)
        for i in range(len(df)):
            local[i] = (local[i]-mean)/len(df)
        return local
        
    
  
        
```

```python
std = Standardization()
df = pd.read_csv('titanic.csv')

print(std.mean_of_list_element(df['Survived']))
std.mean_of_list_elements(df['Survived'])
```

```python
std.median_of_list_elements(df['Survived'])
```
