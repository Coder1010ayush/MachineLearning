# Analyzing data and withdraw useful information - Univariate EDA restaurant dataset


```python
sns.countplot(x='time',data=restaurant)

# conclusion : Number of persons coming for lunch is smaller than dinner in this restaurant
```

```python
sns.countplot(x='sex',data=restaurant)
# conclusion : number of males are more than female 
```

```python
sns.countplot(x='smoker',data=restaurant)
# conclusion : number of smoker coming in the restaurant is less than number of smoker
```

```python
sns.countplot(x='smoker',data=restaurant,hue='sex')
# conclusion : number of male smoker is more than female smoker.
```

```python
print(restaurant['total_bill'].mean())
print(restaurant['tip'].mean())
```

```python
sns.kdeplot(restaurant['total_bill']) #probability distribution function for total_bill
```

```python
sns.kdeplot(restaurant['tip'])
```

```python
# first way to do it through sns.scatterplot method
sns.scatterplot(data=restaurant,x='total_bill',y='tip')
```

```python
# we can see observe above plot wrt another column
sns.scatterplot(data=restaurant,x='total_bill',y='tip',hue='sex')
```

```python
sns.scatterplot(data=restaurant,x='total_bill',y='tip',hue='time')
```

```python
sns.scatterplot(data=restaurant,x='total_bill',y='tip',hue='day')
```

### EDA ---> Titanic dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
titanic = pd.read_csv('titanic.csv')
print(titanic.head())
```

```python
print(titanic.info)
```

```python
print(titanic.shape)
```

```python
sns.countplot(data=titanic,x='Sex')
# conclusion : number of male is greater than female 
```

```python
sns.countplot(data=titanic,x='Survived')
# number of survived person is less than number of death person
```

```python
sns.countplot(data=titanic,x='Pclass')
# conclusion : p3 has more persons than p2 and p1
```

```python
sns.countplot(data=titanic,x='Survived',hue='Sex')
```

```python
sns.countplot(data=titanic,x='Survived',hue='Pclass')
```

```python
sns.countplot(data=titanic,x='Age',hue='Survived')
```

```python
plt.scatter(titanic['Age'],titanic['Survived'])
plt.show()
```

```python
titanic['Survived'].value_counts().plot(kind='pie')
```

```python
titanic['Sex'].value_counts().plot(kind='pie')
```

```python
titanic['Age'].value_counts().plot(kind='pie')
```
